```python
import matplotlib.pyplot as plt 
```

```python
x = [1, 2, 3, 4, 5]   # X-axis values
y = [2, 4, 1, 8, 7]   # Y-axis values
```

```python

plt.plot(x, y)
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=['Saiyam','Ashu','Rajat','Shikha','Yash']
y=[86,99,89,98,97]
```

```python
plt.bar(x,y,color='lightblue' )
plt.title('Marks by Name')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
lables=['python','java','CPP','c']
sizes=[40,20,90,30]
plt.pie(sizes,labels=lables,autopct='%1.0f%%')
plt.legend()
plt.show()
```

```python
data = [25, 30, 30, 45, 50, 55, 60, 65, 70, 75, 80, 90]
```

```python
plt.hist(data,bins=10,color='lightgreen')
plt.title("Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")
plt.show()
```
