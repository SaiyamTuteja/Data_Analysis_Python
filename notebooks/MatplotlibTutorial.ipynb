```python
import matplotlib.pyplot as plt 
import matplotlib.style as st
```

```python
x = [1, 2, 3, 4, 5]   # X-axis values
y = [2, 4, 1, 8, 7]   # Y-axis values
```

```python

plt.plot(x, y)
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=['Saiyam','Ashu','Rajat','Shikha','Yash']
y=[86,99,89,98,97]
```

```python
plt.bar(x,y,color='lightblue' )
plt.title('Marks by Name')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
lables=['python','java','CPP','c']
sizes=[40,20,90,30]
plt.pie(sizes,labels=lables,autopct='%1.0f%%')
plt.legend()
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)  # Normal distribution

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Data')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]


plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = reverse')
plt.legend()
plt.title('Multiple Lines')
plt.show()
```

```python
plt.subplot(1, 2, 1)  # (rows, cols, position)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.bar([1, 2, 3], [3, 2, 5])
plt.title('Plot 2')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')  # Other options: 'seaborn', 'fivethirtyeight', etc.

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title('Styled Plot')
plt.show()
```

```python
plt.style.use('ggplot')
x1=[5,8,10]
y1=[12,16,6]

x2=[6,9,11]
y2=[6,15,7]

plt.plot(x1,y1,'g',label='first Line',linewidth=5)
plt.plot(x2,y2,'c',linewidth=5,label='second line ')

plt.xlabel('xAxis')
plt.ylabel('yAxis')
plt.title('INFO')
plt.grid(False,color='k')
plt.legend()
plt.show()
```
